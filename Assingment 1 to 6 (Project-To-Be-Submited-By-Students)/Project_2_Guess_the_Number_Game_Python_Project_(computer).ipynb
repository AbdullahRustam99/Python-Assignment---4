```python
import random

print("Guess The Number (Computer)")
print("=============================")
while True:
  print("\nThink A Number Between 1 to 50")
  Player_Guess = input("Do You Think Numbr (y/n): ")

  if Player_Guess.lower() == "y":
        num1 = 1
        num2 = 50
        while True:
          Computer_Guess = random.randint(num1,num2)
          print("\nI'm Thinking...")
          print("I Guess it is : ",Computer_Guess)

          print("""
          Too Hight = 1
          Too Low = 2
          Guessed = 3""")
          advice = int(input("Enter Number : "))
          if advice == 1:
            num2 = Computer_Guess
          elif advice == 2:
            num1 = Computer_Guess
          elif advice == 3:
            num1 = 1
            num2 = 50
            print("\n\tYah ! I guessed your number correctly!")
            break
          else:
              print("Invalid Input")
        playAgain = input("\nDo You Want to Play Again ? (y/n) : ")
        if playAgain.lower() == "y":
          continue
        else:
          print("Thank You For Playing")
          break



```
